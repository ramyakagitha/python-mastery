```python
#Transpose a square matrix of n rows and columns
n=int(input("Enter size of matrix:"))
matrix=[]
print("Enter matrix elements:")
for i in range(n):
    row=list(map(int,input().split()))
    matrix.append(row)
print("\n Original Matrix:")
for row in matrix:
    print(row)
print("\nTranspose Matrix:")
for i in range(n):
    for j in range(n):
        print(matrix[j][i],end=" ")
    print()
```

```python
#place even elements of an array at even indexes and odd at odd indexes
arr=list(map(int,input("Enter elements:").split()))
even=[x for x in arr if x%2==0]
odd=[x for x in arr if x%2!=0]
result=[0]*len(arr)
e=o=0
for i in range(len(arr)):
    if i%2==0:
        result[i]=even[e]
        e+=1
    else:
        result[i]=odd[o]
        o+=1
print(result)
```

```python
#check if an array is a subsequence of another
arr1=[1,2,3,4,5]
arr2=[2,4,5]
j=0
for i in arr1:
    if j<len(arr2) and i==arr2[j]:
        j+=1
if j==len(arr2):
    print("Subsequence")
else:
    print("Not a Subsequence")
```

```python
#Unzip nested dicts into expanded dicts
data={
    "A":{"Math":90,"Science":85},
    "B":{"Math":80,"Science":75}
}
result={}
for student,marks in data.items():
    for subject,score in marks.items():
        result[(student,subject)]=score
print(result)
```

```python
#return duplicate elements in sorted order from a list
numbers=list(map(int,input().split()))
dup=[]
for i in numbers:
    if numbers.count(i)>1 and i not in dup:
        dup.append(i)
dup.sort()
print(dup)
```

```python
#sort a list without using the built-in sort method
numbers=list(map(int,input().split()))
n=len(numbers)
for i in range(n):
    for j in range(n-i-1):
        if numbers[j]>numbers[j+1]:
            numbers[j],numbers[j+1]=numbers[j+1],numbers[j]
print(numbers)
```

```python
#find the max number in a list
numbers=list(map(int,input("Enter numbers:").split()))
max=numbers[0]
for i in numbers:
    if i>max:
        max=i
print("Maximum Number=",max)
```

```python
#find the unique number in a list
numbers=list(map(int,input("Enter numbers:").split()))
unique=[]
for i in numbers:
    if numbers.count(i)==1:
        unique.append(i)
print("Unique Elements:",unique)
```

```python
#remove duplicates from a list
numbers=list(map(int,input("Enter numbers:").split()))
result=[]
for i in  numbers:
    if i not in result:
        result.append(i)
print("List after removinng duplicates:",result)
```
